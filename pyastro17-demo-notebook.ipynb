# There are two ways to display matplotlib images

## As static PNGs in the notebook 

To do this, include this special IPython/Jupyter command (called a "magic") before importing matplotlib:

```python
%matplotlib inline
```

## As an interactive plot 

Matplotlib run from the terminal typically includes basic pan/zoom. You can do that in the notebook too, but you should know that if you close the notebook and kernel and then open the notebook later, the interactive plot will have become a static PNG. If you re-run the notebook, the plots will turn back into interactive plots.

To do this, use:

```python
%matplotlib notebook 
```

## Do not try to mix `inline` and `notebook` in the same notebook...

...it may work, but more likely than not you will run into errors or strange behavior (like plots rendering in the wrong cell).



````python
%matplotlib inline
# %matplotlib notebook

import matplotlib.pyplot as plt
````

````python
# These will be used below to set figure sizes...
width = 7 # inches (sorry!)
height = 3
````

````python
# Set the size by calling figure...
plt.figure(figsize=(width, height))
plt.plot([0,10])
````

The cell below will add the new curve to the previous cell if you used `%matplotlib notebook`, but will make a new plot if you used `%matplotlib inline`.

````python
# This will add to the previous cell IF you are using matplotli
plt.plot([1, 5])
````

If you include `plt.figure()` in the cell the plot will always be drawn in the output of that cell.

````python
plt.figure()
plt.plot([0, 100])
````
